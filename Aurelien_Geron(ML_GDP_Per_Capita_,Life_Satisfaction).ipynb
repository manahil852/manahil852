```
# IMPORT LIBRARIES
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
```

```
 # Load the data
df = pd.read_csv("/content/GDP.csv", usecols=[0, 1, 2])
```

```
def prepare_country_stats(df):
    # Drop any rows with missing values (optional)
    df = df.dropna()

    # Select only the relevant columns
    df = df[["COUNTRY", "GDP per Capita", "Life Satisfaction"]]

    # Sort by GDP per Capita (optional, but good for visualization)
    df = df.sort_values(by="GDP per Capita")

    # Set country as index (optional)
    df.set_index("COUNTRY", inplace=True)

    return df
```

```
# Prepare the data
country_stats = prepare_country_stats(df)
X = np.c_[country_stats["GDP per Capita"]]
y = np.c_[country_stats["Life Satisfaction"]]
```

```
country_stats.plot(kind='scatter', x="GDP per Capita", y='Life Satisfaction')
plt.show()
```

```
 # Select a linear model
model = sklearn.linear_model.LinearRegression()
```

```
# Remove commas and convert to float
df["GDP per Capita"] = df["GDP per Capita"].str.replace(",", "").astype(float)
```

```
X = np.c_[df["GDP per Capita"]]
y = np.c_[df["Life Satisfaction"]]
```

```
# Train the model
model.fit(X,y)
```

```
# Make a prediction for Cyprus
X_new = [[22587]]  # Cyprus' GDP per capita
print(model.predict(X_new)) # outputs [[ 5.96242338]]

```

```
# Make a prediction for pakistan
Y_new  = [[1365]]  # Pakistan GDP per capita
print(model.predict(Y_new)) # outputs [[4.35175632]]
```
